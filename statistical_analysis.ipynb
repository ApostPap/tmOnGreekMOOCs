```python
# Print the number of agent targeted and team interventions

#C - Targeted Interventions
#E - Team Interventions
def getAgentInterventions(mooc, show_all=False):
    
    print("\nPrint the number of agent targeted and team interventions\n") 
    
    session_target_int = 0
    session_team_int = 0
    total_targeted_int = 0
    total_team_int = 0

    for index, row in mooc.iterrows():
      #  message_counter = message_counter + 1
        message_type = row['message_type']
        
        if(message_type==9 or message_type==11):
            session_target_int = session_target_int+1
            total_targeted_int = total_targeted_int + 1
        if(message_type==6 or message_type==7 or message_type==12):
            total_team_int = total_team_int +1
            session_team_int =session_team_int + 1
            
        
        if(row['session_id'] >0):
            message_counter = 0

            #print("For session : ",str(row['session_id']))
            #print("Agent's Last intervention after: "+ str(last_interv_sec) +"sec at: " + str(last_interv_timestamp)+ " after "+ str(last_interv_message_counter-2)+" messages")
            if(show_all):
                print("Session: "+str(int(row['session_id']))+", Team Intervations: "+str(session_team_int)+", Targeted Intervations: "+str(session_target_int))
            #print(str(session_team_int+session_target_int))

           
            session_team_int = 0
            session_target_int = 0
    
    
    print("Total Targeted Interventions: "+ str(total_targeted_int))
    print("Total Team Interventions: "+ str(total_team_int))

    
    return
```

```python
# Print average intervation answer length (not including final answer)

# G - Average Intervention Answer Length
# H - Average Sec between Intervention & Answer 

def getAgentInterventionAnswersLen(mooc, show_all=False):
    
    
    print("\nPrint average intervation answer length (not including final answer)\n")
    session_target_int = 0
    session_team_int = 0
    total_targeted_int = 0
    total_team_int = 0
    session_answer_len = 0
    session_answers = 0

    for index, row in mooc.iterrows():
      #  message_counter = message_counter + 1
        message_type = row['message_type']
        
        if(row['feedback_reason'] == 0):
            session_answer_len = session_answer_len + row['message_length']
            session_answers = session_answers+1
            
        
        if(row['session_id'] >0):
            message_counter = 0

            #print("For session : ",str(row['session_id']))
            #print("Agent's Last intervention after: "+ str(last_interv_sec) +"sec at: " + str(last_interv_timestamp)+ " after "+ str(last_interv_message_counter-2)+" messages")
            if(show_all):
                if(session_answers>0):
                #print(str(int(row['session_id']))+","+str('{0:.2f}'.format(session_answer_len/session_answers)))
                    print("Session: "+str(int(row['session_id']))+", Average Intervation Answer length: "+str('{0:.2f}'.format(session_answer_len/session_answers)))
                else:
                #print(str(int(row['session_id']))+",0")
                    print("Session: "+str(int(row['session_id']))+", Average Intervation Answer length: 0")


           # session_counter = session_counter + 1
            session_answer_len = 0
            session_answers = 0
    
    
    
    
    
    answers = mooc.loc[mooc['feedback_reason'] == 0]
    sum_message_len = answers['message_length'].sum()
    print("Total Answers: "+str(len(answers.index)))
    averge_answer_len = (sum_message_len/len(answers.index))
    print("Average Answer Length: "+str('{0:.2f}'.format(averge_answer_len)))
    return 
```

```python
# Print average final answer length

#Q - Final Answer Length 
def getFinalAnswersLen(mooc, show_all = False):
    
    print("\nPrint average final answer length\n")
   
    sum_session_answer_len = 0
    session_answers = 0
    
    session_answer_len = 0

    for index, row in mooc.iterrows():
      #  message_counter = message_counter + 1
        message_type = row['message_type']
        
        if(row['message_type'] == 0 and int(row['message_length']) > 0):
            
            session_answer_len = row['message_length']
            sum_session_answer_len = sum_session_answer_len + row['message_length']
            session_answers = session_answers+1
        if(row['session_id'] >0):
            message_counter = 0
            #print(str(int(row['session_id'])))
            if(show_all):
                print("Session: "+str(int(row['session_id']))+", Final answer length: "+str(session_answer_len))
    
    
    
    
    #Some times we have errors in the dataset and it shows empty answers while we have the actuall answer at the top
    answers = mooc.loc[(mooc['message_type'] == 0) &(len(mooc['the_message']) > 0)]
    sum_message_len = answers['message_length'].sum()
    print("Total Final Answers: "+str(len(answers.index)))
    averge_answer_len = (sum_session_answer_len/len(answers.index))
    print("Average Final Answer Length: "+str('{0:.2f}'.format(averge_answer_len)))
    return 
```

```python
#Prints sec until the latest message before final answer

#R - Duration Until Final Answer in Seconds
def getDurationBeforeFinalAnswer(mooc,show_all=False):
    
    print("\nPrints seconds until the latest message before final answer\n")
    previous_sec=-1
    count = 0
    sum_sec = 0
    for index, row in mooc.iterrows():
        message_type = row['message_type']
        if(index>0):
            if(row['session_id'] >0):
                count = count + 1
                sum_sec = sum_sec + previous_sec

                if(show_all):
                    print("Session: "+str(int(row['session_id']))+", Final answer length: "+str(previous_sec))
            if(message_type!=6):
                previous_sec = row['sec']
        else:
            previous_sec = row['sec']
    if(count>0):
        print("Average duration before finding final answer: "+ str('{0:.2f}'.format(sum_sec/count)))
```

```python
#Prints the average time and messages between intervention

#Its message_counter-2 because we have the final answer at the top
def getAgentInterventionStats(mooc,show_all = False):
    
    print("\nPrints the average time and messages between intervention\n")
    divide_counter=0
    sum_sec_dif = 0
    sum_message_dif = 0
    session_counter = 1
    message_counter = 0
    last_interv_message_counter = 0
    last_interv_sec = 0
    start_counting = False

    for index, row in mooc.iterrows():
        message_counter = message_counter + 1
        message_type = row['message_type']
        
        if(row['session_id'] >0):
            message_counter = 0
            start_counting = False
            session_counter = session_counter + 1
            if(show_all):
                print("For session number: ",str(session_counter))
            
        else:
        
            if(message_type==6 or message_type==7 or message_type==9 or message_type==11 or message_type==12):
                if(start_counting):
                    divide_counter = divide_counter+1
                    sum_sec_dif = sum_sec_dif + (row['sec'] - last_interv_sec)
                    sum_message_dif = sum_message_dif + message_counter
                    if(show_all):
                        print("Next intervention: ")
                        print("Messages until next: ",str(message_counter))
                        print("Diff: ",str(last_interv_sec), " - ", str(row['sec']), " sec")
                
                else:
                    start_counting = True
                    
                last_interv_sec = row['sec']
                message_counter = 1
        
        
    if(divide_counter>0):
        print("Average time duration between interventions: "+ str('{0:.2f}'.format(sum_sec_dif/divide_counter)))
        print("Average messages count between interventions: "+ str('{0:.2f}'.format(sum_message_dif/divide_counter)))
```

```python
#Prints first agent intervention per session showing the timestamp and the number of messages until the intervention

#K - First Agent Intervention At Second
#L - First Agent Intervention Timestamp
#M - Posts until First Agent Intervention

#Its message_counter-2 because we have the final answer at the top
def getFirstAgentIntervention(mooc):
    
    print("\nPrints first agent intervention per session showing the timestamp and the number of messages until the intervention\n")
    session_counter = 1
    found_first_interv = False
    message_counter = 0
    this_session_timestamp = ""
    this_session_sec = 0
    this_session_message_number = 0
    for index, row in mooc.iterrows():
        message_counter = message_counter + 1
        message_type = row['message_type']
        if(not(found_first_interv) and (message_type==6 or message_type==7 or message_type==9 or message_type==11 or message_type==12)):
           # print("For session number: ",str(session_counter))
           # print("Agent's first intervention after: "+ str(row['sec']) +"sec at: " + str(row['timestamp'])+ " after "+ str(message_counter-2)+" messages")
            
            #print(str(row['sec'])+","+str(row['timestamp'])+","+str(message_counter-2))
            
            this_session_sec = row['sec']
            this_session_timestamp = row['timestamp']
            this_session_message_number = message_counter-2
            
            found_first_interv = True
            message_counter = 0
        
        if(row['session_id'] >0):   
            if(found_first_interv==False):
                print("Session: "+str(int(row['session_id']))+", No intervention found")
            else:
                print("Session: "+str(int(row['session_id']))+", Seconds: "+str(this_session_sec)+", Timestamp: "+str(this_session_timestamp)+", Message #: "+str(this_session_message_number))

            found_first_interv = False
            message_counter = 0

            session_counter = session_counter + 1

    
```

```python
#Prints last agent intervention per session showing the timestamp and the number of messages until the intervention

#Its message_counter-2 because we have the final answer at the top

#N - Last Agent Intervention At Second
#O - Last Agent Intervention Timestamp
#P - # Posts until Last Agent Intervention
def getLastAgentIntervention(mooc):
    
    print("\nPrints last agent intervention per session showing the timestamp and the number of messages until the intervention\n")
    session_counter = 1
    message_counter = 0
    last_interv_message_counter = 0
    last_interv_sec = 0
    last_interv_timestamp = 0
    agent_int_found = False

    for index, row in mooc.iterrows():
        message_counter = message_counter + 1
        message_type = row['message_type']
        if(message_type==7 or message_type==9 or message_type==11 or message_type==12):
            agent_int_found = True
            last_interv_sec = row['sec']
            last_interv_timestamp = row['timestamp']
            last_interv_message_counter = message_counter
        
        if(row['session_id'] >0):
            message_counter = 0

            #print("For session number: ",str(session_counter))
            #print("Agent's Last intervention after: "+ str(last_interv_sec) +"sec at: " + str(last_interv_timestamp)+ " after "+ str(last_interv_message_counter-2)+" messages")
            if(agent_int_found):
                
                print("Session: "+str(int(row['session_id']))+", Seconds:"+str(last_interv_sec)+", Timestamp: "+str(last_interv_timestamp)+", Message #: "+str(last_interv_message_counter-2))
            else:
                print("Session: "+str(int(row['session_id']))+", No intervention found")
                #print("0,-,0")
            agent_int_found = False

            last_interv_sec = 0
            last_interv_timestamp = 0
            last_interv_message_counter = 0
            session_counter = session_counter + 1

    
```

```python
# Prints when an intervention is answered (if show_all = true) showing intervention timestamp and answer timestamp.
# At the end it prints "total interventions", "targeted answered interventions", "team answered interventions", "not answered interventions"
 

#Its message_counter-2 because we have the final answer at the top

#D - Targeted Interventions Answered
#F - Team Interventions Answered
#H - Average Sec between Intervention & Answer

def getInterventionAnswerStats(mooc,keep_only_first_answer = True,show_all = False):
    
    print("\nPrints when an intervention is answered (if show_all = true) showing intervention timestamp and answer timestamp.\n")
    intervention_counter = 0
    intervention_answered_targeted = 0
    intervention_answered_team = 0

    session_counter = 0
    session_interventions_messages = []
    session_interventions_type = []
    session_interventions_timestamps = []
    session_interventions_sec = []
    session_question_ids = [] 
    session_answer_timestamp = []
    session_answer_sec = []
    session_interv_team_answered = 0
    session_interv_targeted_answered = 0
    session_answer_sec_sum = 0

    
    
    for index, row in mooc.iterrows():
        message_type = row['message_type']
        
        if(row['session_id'] >0):
            session_answer_sec_sum = 0

            session_interv_team_answered = 0
            session_interv_targeted_answered = 0
            if(show_all):
                print("For session number: ",str(session_counter))
            interv_index = 0
            for intervention in session_interventions_messages:
                answer_index = 0
                for answer in session_question_ids:
                    if(intervention == answer ):

                        if(show_all):
                            
                            print("Intervention at: ", str(session_interventions_sec[interv_index]))
                            print("Answer at: ", str(session_answer_sec[answer_index]))
                        
                        session_answer_sec_sum = session_answer_sec_sum+(int(session_answer_sec[answer_index]) - int(session_interventions_sec[interv_index]))

                            
                        
                        if(session_interventions_type[interv_index] == 7 or session_interventions_type[interv_index] == 12):
                            intervention_answered_team = intervention_answered_team + 1
                            session_interv_team_answered = session_interv_team_answered + 1

                        else: 
                            intervention_answered_targeted = intervention_answered_targeted + 1
                            session_interv_targeted_answered = session_interv_targeted_answered+1

                        if(keep_only_first_answer):
                            break
                    answer_index = answer_index+1
                
                interv_index = interv_index + 1
            
            
            
            
            session_interventions_messages = []
            session_interventions_timestamps = []
            session_interventions_sec = []
            session_question_ids = [] 
            session_answer_timestamp = []
            session_interventions_type = []
            session_answer_sec = []
            
            session_counter = session_counter + 1
            
            #print("Team ans "+str(session_interv_team_answered))
            #print("Target ans "+str(session_interv_targeted_answered))
            #print("s" + str(session_answer_sec_sum))
            if(show_all):
                if(session_interv_team_answered==0 and session_interv_targeted_answered==0):
                    print("Session: "+str(int(row['session_id']))+", Session Team Answer #: "+str(session_interv_team_answered)+", Session Targeted Answer #: "+str(session_interv_targeted_answered)+", Average Dutation betweem Intervation & Answer: "+str("0"))
                else:
                    print("Session: "+str(int(row['session_id']))+", Session Team Answer #: "+str(session_interv_team_answered)+", Session Targeted Answer #: "+str(session_interv_targeted_answered)+", Average Dutation betweem Intervation & Answer: "+str('{0:.2f}'.format(session_answer_sec_sum/(session_interv_team_answered+session_interv_targeted_answered))))

            
        else:
        
            if(message_type==6 or message_type==7 or message_type==9 or message_type==11 or message_type==12):
                intervention_counter = intervention_counter + 1
                session_interventions_type.append(message_type)
                session_interventions_messages.append(row['the_message'])
                session_interventions_timestamps.append(row['timestamp'])
                session_interventions_sec.append(row['sec'])
        
                
            elif(message_type==10):              
                session_question_ids.append(row['question_id'])
                session_answer_timestamp.append(row['timestamp'])
                session_answer_sec.append(row['sec'])

                    
                
        
    print("Total Intervetions: ", str(intervention_counter))
    print("Total Targeted Answers: ", str(intervention_answered_team))
    print("Total Team Answers: ", str(intervention_answered_targeted))
    print("Total Not answered: ",str(intervention_counter - intervention_answered_team - intervention_answered_targeted))
    
```
